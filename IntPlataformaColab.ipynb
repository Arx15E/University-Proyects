0. Cargar librerias

```python
import numpy as np #funciones numericas
import pandas as pd #base de datos

```

1. conexion con la base de datos

```python
nxl = "/content/0. SolicitantesCrédito(Quiz 1).xlsx"
XDB = pd.read_excel(nxl,sheet_name=0) #Dataframe, datos - word
XDB.head(10)
```

2. llamar la variable de caracterización (Edad)

```python
XDe = XDB["Edad"]
print(XDe)
XDe=XDe.dropna() #Evitar celdas vacias
XDe=np.array(XDe)
print(XDe)
```

4. se calcula la media y la desviacion estandar de la edad

```python
print("La media de la edad es: ", np.mean(XDe))
print("La desviacion estandar de la edad es: ", np.std(XDe))
print("La eadad minima es: ", np.mean(XDe)-2*np.std(XDe))
print("La eadad maxima es: ", np.mean(XDe)+2*np.std(XDe))
```

5. histograma de frecuencias - agrupar los datos por intervalos

```python
import matplotlib.pyplot as plt
plt.hist(XDe,bins=20)
plt.show()
#
```

6. calcular el valor de pertenencia para una persona de 60 años (0,1) - afinidad de un valor por un variable aleatoria

```python
VP=np.exp(-0.5*((49.12-np.mean(XDe))/np.std(XDe))**2)
print("el valor de pertenencia de una edad de 60 años es: ",VP)
```
